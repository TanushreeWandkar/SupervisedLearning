### Linear Regression

Linear regression is a type of statistical analysis used to predict the relationship between two variables. It assumes a linear relationship between the independent variable and the dependent variable, and aims to find the best-fitting line that describes the relationship. The line is determined by minimizing the sum of the squared differences between the predicted values and the actual values.

Linear regression shows the linear relationship between the independent(predictor) variable i.e. X-axis and the dependent(output) variable i.e. Y-axis.  If there is a single input variable X(independent variable), such linear regression is called simple linear regression.

This algorithm explains the linear relationship between the dependent(output) variable y and the independent(predictor) variable X using a straight line  Y= B0 + B1 X or y=mx+c

Goal of Linear Regression : The goal of the linear regression algorithm is to get the best values for B0 and B1 to find the best fit line. The best fit line is a line that has the least error which means the error between predicted values and actual values should be minimum.

### Loss Function : Error of single training sample. ie difference between Actual and Predicted values.
### Cost Function : Average of loss function over an entire dataset

### How do we find a best fit line: 
The cost function helps to work out the optimal values for B0 and B1. In Linear Regression, generally Mean Squared Error (MSE) cost function is used. Using the MSE function, well update the values of B0 and B1 such that the MSE value settles at the minima.  These parameters can be determined using the gradient descent method such that the value for the cost function is minimum.



### Gradient Descent for Linear Regression
Gradient Descent is one of the optimization algorithms that optimize the cost function(objective function) to reach the optimal minimal solution. To find the optimum solution we need to reduce the cost function(MSE) for all data points. This is done by updating the values of B0 and B1 iteratively until we get an optimal solution.
for example: lets take that you are at the top of an U-shaped pit and there are only limited number of steps you can take to reach the bottom of the pit. If you take larger steps, you might overshoot the bottom, on the other hand if you take one step at a time, you might take a longer time to reach the bottom. How big of a step you take is called Learning rate (aplha) which determines how fast the algorithm converges to the minima.



### Evaluation Metrics for Regression:
1. R-sqaured : R-Squared is a number that explains the amount of variation that is explained/captured by the developed model. It always ranges between 0 & 1 . Overall, the higher the value of R-squared, the better the model fits the data.

R2 = 1  ( RSS/TSS ) 

RSS : Residual sum of sqaures.Its the difference between the expected and the actual observed output.



TSS : Total sum of sqaures. Sum of errors of the data points from the mean of the response variable.



2. Root mean sqaure error (RMSE): root of mean sqaured error.


### Assumptions of Linear Regression
LR is a parametric approach meaning it makes some assumptions about the data in its process.Below are the asssumptions.
1. Linearity in the residuals: the dependent and independant variables should be linear.
2. Independance of residuals/ Autocorrelation : There should not be any kind of correlation/ relationship / visible patterns in the error terms/residuals. No autocorrelation.
3. Normal distribution of the residuals: The residuals should be normally distributed. If the error terms are non-normally distributed, suggests that there are a few unusual data points that must be studied closely to make a better model.
4. The equal variance of the residuals:

### Considerations of Multiple Linear Regression
Along with the general assumptions of the simple LR, there are some additional ones for MLR.

Overfitting: When more and more variables are added to a model, the model may become far too complex and usually ends up memorizing all the data points in the training set. This phenomenon is known as the overfitting of a model. This usually leads to high training accuracy and very low test accuracy.

Multicollinearity: It is the phenomenon where a model with several independent variables, may have some variables interrelated.
Detect multicollinearity by using VIF. VIF basically explains the relationship of one independent variable with all the other independent variables.if VIF > 10 then the value is definitely high and it should be dropped. And if the VIF=5 then it may be valid but should be inspected first. If VIF < 5, then it is considered a good vif value.

Feature Selection: With more variables present, selecting the optimal set of predictors from the pool of given features (many of which might be redundant) becomes an important task for building a relevant and better model.
